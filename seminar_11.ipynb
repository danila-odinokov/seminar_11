f(x) = -12x^4*sin(cos(x)) - 18x^3+5x^2 + 10x - 30

2. Найти интервалы, на которых функция возрастает

3. Найти интервалы, на которых функция убывает

4. Построить график

5. Вычислить вершину

6. Определить промежутки, на котором f > 0

7. Определить промежутки, на котором f < 0

```python
from sympy import *
from sympy.plotting import plot
init_printing()
```

*1. Определить корни*

```python
x = Symbol('x')
f = -5*x**2+10*x-150
solve(f, x)              # решение уравнения
```

*2. Найти интервалы, на которых функция возрастает*

```python
der_x = diff(f)                              # производная функции
solve_poly_inequality( Poly(der_x), '>')     # решаем неравенство вида (производная < 0), чтобы найти интервал возрастания
```

*3. Найти интервалы, на которых функция убывает*

```python
solve_poly_inequality( Poly(der_x), '<')     # решаем неравенство вида (производная < 0), чтобы найти интервал убывания
```

*4. Построить график*

```python
plot(f, (x, -15, 15))
```

*5. Вычислить вершину*

```python
print(f' Производная функции = {der_x}')
```

**Производная функции представляет собой уравнение вида 2a*x + b**

**Приравниваем нашу производную к нулю**

**-10*x + 10 = 0**

**Значение a = -5, значение b = 10**

```python
a = -5
b = 10
x_high = -b / (2*a)
print(x_high, f.subs(x, x_high))   # вершина параболы
```

*6. Определить промежутки, на котором f > 0*

```python
solve_poly_inequality( Poly(f), '>') 
```

*7. Определить промежутки, на котором f < 0*

```python
solve_poly_inequality( Poly(f), '<') 
```
